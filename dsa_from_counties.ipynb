```python
import csv

#read in dsa-county mapping
filename = "data/dsa_counties_revised.csv"
reader = csv.DictReader(open(filename, 'rU'))

#dsas with their county components
dsa_dict = {}

#fips and dsa pair with fips as the key
fips_dict = {}

#data results
dsa_data_dict = {}

for row in reader:

    fips = row["FIPS"].replace("'", "")
    dsa = row["DSA"]
    if dsa in dsa_dict:
        
        fips_list = dsa_dict[dsa]
        fips_list.append(fips)
        dsa_dict[dsa] = fips_list
    else:
        dsa_dict[dsa] = [fips]
        
        
    fips_dict[fips] = dsa
    
        
    
#read in data file
rowcount = 0
filename = "data/ACS_13_1YR_B27001_with_ann.csv"
data_reader = csv.DictReader(open(filename, 'rU'))

#write out initial rows
writer = csv.writer(open("dsa_insurance_data.csv", 'wb'))
writer.writerow(["DSA", "TOTAL POP ESTIMATE", "Male: - 18 to 24 years: - No health insurance coverage"])


for row in data_reader:
    #save first line
    rowcount = rowcount + 1
    if rowcount == 1:
        continue
    
    rowcount = rowcount + 1
    
    geo_id = row["GEO.id2"]
    while len(geo_id) < 5:
        geo_id = "0" + geo_id
    
    #get dsa
    if geo_id in fips_dict:
        this_dsa = fips_dict[geo_id]
        #print this_dsa
        
        if this_dsa in dsa_data_dict:
            dsa_data = dsa_data_dict[this_dsa]
            total = dsa_data["HD01_VD01"] + int(row["HD01_VD01"])
            from18to24 = dsa_data["HD01_VD11"] + int(row["HD01_VD11"])
            dsa_data_dict[this_dsa] = {"HD01_VD01": total, "HD01_VD11": from18to24}
        else:
            dsa_data_dict[this_dsa] = {"HD01_VD01": int(row["HD01_VD01"]), "HD01_VD11": int(row["HD01_VD11"])}


sorted_dsas = sorted(dsa_data_dict.keys())
#print sorted_dsas

for dsa in sorted_dsas:
    dsa_data = dsa_data_dict[dsa]
    writer.writerow([dsa, dsa_data["HD01_VD01"], dsa_data["HD01_VD11"]])
```
